# Visualization Part :
Using Plots to check if the variables fit the theoretical normal or exponential distribution.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Setting the dark background theme for the plots
sns.set(style="darkgrid")
plt.style.use("dark_background")

# Creating distribution plots for each numerical variable to check for normal or exponential distribution fit
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(14, 22))
fig.tight_layout(pad=5.0)

for i, var in enumerate(numerical_vars):
    sns.histplot(data[var], kde=True, ax=axes[i//2, i%2], color="skyblue").set_title(var)

plt.show()
```

### _Distribution Insights_:

* __Customer Lifetime Value & Total Claim Amount__: Show right-skewed distributions, hinting at exponential characteristics. A large number of customers have values on the lower end, with fewer customers having very high values.

* __Income__: Also right-skewed, indicating many customers have low income, with a spike at the $0 income mark, possibly indicating unemployed or non-reporting customers.

* __Monthly Premium Auto__: Slightly right-skewed, suggesting most customers pay lower premiums, with fewer customers at higher premiums.

* __Months Since Last Claim & Months Since Policy Inception__: These don't fit neatly into normal or exponential distributions, showing more uniform or bimodal tendencies.

* __Number of Open Complaints__: Highly concentrated at 0, indicating most customers have no complaints.

* __Number of Policies__: Shows spikes at certain policy counts, not fitting a clear normal or exponential trend.

# Plotting using Log transformation
Using log transformation to see if it changes the result.

```python
import numpy as np

# Applying log transformation where applicable (excluding columns with 0 values to avoid -inf)
log_transformable_vars = ['Customer Lifetime Value', 'Monthly Premium Auto', 'Total Claim Amount']
data_log_transformed = data.copy()

# Applying the log transformation
for var in log_transformable_vars:
    data_log_transformed[var] = np.log(data_log_transformed[var])

# Creating distribution plots for the log-transformed variables
fig, axes = plt.subplots(nrows=len(log_transformable_vars), ncols=1, figsize=(10, 15))
fig.tight_layout(pad=5.0)

for i, var in enumerate(log_transformable_vars):
    sns.histplot(data_log_transformed[var], kde=True, ax=axes[i], color="skyblue").set_title(f'Log-transformed {var}')

plt.show()
```

### _Log Transformation Insights:_
* __Customer Lifetime Value__: The log transformation makes the distribution more bell-shaped, indicating a better fit for normal  distribution assumptions after the transformation.

* __Monthly Premium Auto__: Similarly, becomes more normally distributed, with the skewness reduced significantly, showcasing the power of log transformation in dealing with right-skewed data.

* __Total Claim Amount__: Also shows improvement towards a normal distribution, though still slightly skewed, the transformation helps in making the data more symmetrical.
